Funciones = analizan pedazos de código que se les da para devolverlos o hacer cosas con ellos

DIR = Devuelve la lista de atributos válidos del objeto pasado.

UPPER = Convierte a mayúsculas.

LOWER = Convierte a minúsculas.

CAPITALIZE = Primera en mayúscula

FIND = Método encuentra la primera aparición del valor especificado, sino devuelve 1.

INDEX = Método encuentra la primera aparición del valor especificado, sino devuelve una excepción

ISNUMERIC = Si es númerico devuelve True.

ISALPHA = Si es alfa númerico devuelve True.

COUNT = Devuelve el número de ocurrencias de una subcadena en la cadena dada.

LEN = Cuenta los caracteres de una cadena.

ENDSWITH = Verifica si una cadena empieza con.

STARTSWITH = Verifica si una cadena termina con.

REPLACE = Remplaza un valor por otro.

SPLIT = Separa por el parametro dado.

```python
cadena1 = "Hola soy mega"
cadena2 = "Bienvenido amigo"
# Variables las cuales seran utilizadas para buscar metodos
```

```python
resultado = cadena1.upper()
print(resultado) # Son retornadores que transforman los simbolos dependiendo de el metodo que se les imponga
```

```python
busqueda_find = cadena1.find("hola")
print(busqueda_find) # Busqueda que busca coincidencias y si no existen suelta resultado -1
```

```python
busqueda_find = cadena1.index("x")
print(busqueda_find) # Busqueda que busca numero de coincidencias y si no existen suelta un error y para el programa
```

```python
busqueda_find = cadena1.count("x")
print(busqueda_find)# Busqueda que busca numero de coincidencias y si no existen marca 0 sin dar error y terminar el programa
```

```python
contar_caracteres = len(cadena1)
print(contar_caracteres) # cuenta los caracteres de la variable incluyendo los espacion como un caracter
```

```python
cadena_nueva = cadena1.replace("la" , "lu")
print(cadena_nueva) # Remplaza los caracteres por los nuevos
```
